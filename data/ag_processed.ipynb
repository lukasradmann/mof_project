```python
import pandas as pd
import pprint
from IPython.display import display

df = pd.read_csv('../data/processed_MOFs.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test = train_test_split(df, test_size=0.2, random_state=42)

# X_train, y_train are the training data
# X_test, y_test are the testing data
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
train_data = TabularDataset(X_train)
sample_size = 30000
train_data = train_data.sample(n=sample_size, random_state=42)
label = 'CO2_uptake_P0.15bar_T298K [mmol/g]'
predictor = TabularPredictor(label=label).fit(train_data, time_limit=60)
```

```python
print("AutoGluon infers problem type is: ", predictor.problem_type)
print("AutoGluon identified the following types of features:")
print(predictor.feature_metadata)
```

```python
test_data = TabularDataset(X_test)
y_test = test_data[label]
test_data = test_data.drop(columns=[label], axis=1)
test_data
```

```python


y_pred_proba = predictor.predict(test_data)
y_pred_proba.head()
```

```python
predictor.evaluate(test_data)
```

```python
performance = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred_proba)
performance
```

```python
predictor.leaderboard(test_data)
```

```python
predictor.feature_importance(test_data)
```
